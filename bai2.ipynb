Xây dựng mô hình Softmax Regression

```python
import numpy as np
from tqdm import tqdm

class SoftmaxRegression:
    def __init__(self, epoch: int, lr: float) -> None:
        self.epoch = epoch
        self.lr = lr
        self.losses = []
        self.metrics = []

    def softmax(self, z: np.ndarray) -> np.ndarray:
        exp_z = np.exp(z - np.max(z, axis=-1, keepdims=True))  
        return exp_z / np.sum(exp_z, axis=-1, keepdims=True)

    def loss_fn(self, y: np.ndarray, y_hat: np.ndarray) -> float:
        return -np.mean(np.sum(y * np.log(y_hat + 1e-6), axis=-1))  

    def accuracy(self, y: np.ndarray, y_hat: np.ndarray) -> float:
        return np.mean(y.argmax(axis=-1) == y_hat.argmax(axis=-1))  

    def fit(self, X: np.ndarray, y: np.ndarray) -> None:
        n, d = X.shape
        n_classes = y.shape[-1]
        self.theta = np.zeros((d, n_classes))

        with tqdm(range(self.epoch)) as pb:
            for e in pb:
                pb.set_description(f"Epoch {e+1}")

                s = self.predict(X)
                s_c = s - y
                gradient = np.matmul(X.T, s_c) / n if n > 0 else 0 
                self.theta -= self.lr * gradient

                s = self.predict(X)
                loss = self.loss_fn(y, s)
                acc = self.accuracy(y, s)

                pb.set_postfix({"loss": loss, "acc": acc})
                self.losses.append(loss)
                self.metrics.append(acc)

    def predict(self, X: np.ndarray) -> np.ndarray:
        z = np.matmul(X, self.theta)
        y_hat = self.softmax(z)
        return y_hat
```

Khởi tạo và huấn luyện mô hình

```python
softmax_reg = SoftmaxRegression(200, 0.0001)
```

```python
softmax_reg.fit(X, oh_y)
```

```python
x_axis = np.array(range(200))
plt.scatter(x_axis, softmax_reg.losses, c="r", label ="loss")
```

```python
plt.scatter(x_axis, softmax_reg.metrics , c="b", label = "accuracy")
```

```python
x_axis = np.array(range(200))
plt.figure(figsize=(10, 5))
plt.plot(x_axis, softmax_reg.losses, c="r", label="Loss")
plt.plot(x_axis, softmax_reg.metrics, c="b", label="Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Giá trị")
plt.legend()
plt.title("Loss và Accuracy trong quá trình huấn luyện")
plt.show()
```
